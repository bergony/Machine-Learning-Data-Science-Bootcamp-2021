## Two options for subPlots

```python
# Subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                        ncols=2,
                                        figsize=(10,5))

#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python

# subplt option 2

fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
#plt to each different index


ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one Should you see? (Pyplot vs Mathplotlib OO Method)
* When plotting something quickly okay to use pyplot method
* When plotting something quickly more advanced use  the OO method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
#Pyplot Method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c="target",
            figsize=(10,6))
```

```python
# OO Method mix wih pyplot
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
#ax.set_xlim([45,100]);
```

```python
# OO method from scrach

fig , ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
# Customize the plot
ax.set(title="Heart Disease and cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");
#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add horizontal Line
ax.axhline(over_50['chol'].mean(),
          linestyle='--')
```

```python
over_50.head()
```

```python
# Subplot of chol, age , thalch
fig , (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex = True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"]);

# Consutmizer ax0
ax0.set(title="Heart Diasease and cholestorerol Levels", ylabel="Cholestoterol");

# add legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add Horizonta line
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");

#add Data to ax1

scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"]);

# Costomize ax1

ax1.set(title="Heart Disease and Thalach Levels", xlabel="Age", ylabel="Thalach")

# Add Legend

ax1.legend(*scatter.legend_elements(), title="Target")

# Add Horizonta Line

ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

# Add title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
plt.style.use('seaborn-whitegrid')

# OO method from scrach
fig , ax = plt.subplots(figsize=(10,6))

#Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter"); #This change color schema
# Customize the plot
ax.set(title="Heart Disease and cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");
#Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

#Add horizontal Line
ax.axhline(over_50['chol'].mean(),
          linestyle='--');
```

```python
# Customizing the y and x axis limitations
# Subplot of chol, age , thalch
fig , (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10,10), sharex = True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter");

# Consutmizer ax0
ax0.set(title="Heart Diasease and cholestorerol Levels", ylabel="Cholestoterol");

#Change the x axis limint
ax0.set_xlim([50,80]);

# add legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add Horizonta line
ax0.axhline(y=over_50["chol"].mean(), linestyle="--");

#add Data to ax1

scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"], cmap="winter");

# Costomize ax1

ax1.set(title="Heart Disease and Thalach Levels", xlabel="Age", ylabel="Thalach")

#Change ax1 x limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add Legend

ax1.legend(*scatter.legend_elements(), title="Target")

# Add Horizonta Line

ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

# Add title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
