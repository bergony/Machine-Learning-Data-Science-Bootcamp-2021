## Matplotlib exemple workflow

```python
# 0. import mathplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10,10)) #width , hight

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show( you save the whole figure)
fig.savefig("images/sample-plot.png")




```


## Making figure with NUmpy arrays

We Want:
* Line Plot
* Scatter plit
* Bar Plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some Data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create a line Plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same date to make a scatter

fig, ax= plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
#Another scatter plot
fig, ax = plt.subplots()

ax.scatter(x, np.sin(x));
```

```python
#make a plot from a dictonary
nut_butter_prices = {"Almond butter": 10,
                    "peanut": 8,
                    "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Dan's Nut Butter Score", 
      ylabel="Price ($)");
```

```python
fig , ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

## Two options for subPlots

```python
# Subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                        ncols=2,
                                        figsize=(10,5))

#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python

# subplt option 2

fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
#plt to each different index


ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```
